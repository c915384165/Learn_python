# 文件的使用

## 文件的类型

文件的概念：

文件是数据的抽象和集合

- 文件是存储在辅助存储器上的数据序列
- 文件是数据存储的一种形式
- 文件的展现形态：文本文件和二进制文件

文本文件vs二进制文件

- 本质相同，都是二进制文件；
- 展示形式不同，编码不同。
- utf-8天然的文本格式；
- png还有其他格式，

f.txt 文件保存：“中国是一个伟大的国家“

```python
#文本形式打开文件
tf = open("f.txt", "rt") # open(): function open one file, parameters: filename; r: read, t: text mode
print(tf.readline()) # .readline() read one line
tf.close() # close file
```

```python
#二进制形式打开文件
tf = open("f.txt", "rb") # b: bin mode
print(tf.readline())
tf.close()
```

## 文件内容的读取

打开文件后可以：数据读入和数据输出 即：读文件 写文件

### 读文件函数

- a.read(size)
- a.readline(size)
- a.readlines(hint)

### 遍历文件

```python
fname = input("请输入打开文件的名称")
```

```python
fo = open(fname, 'r')
txt = fo.read(2)
fo.seek(0)
while txt != "":
    txt = fo.read(2)
    print(txt)
fo.close()
```

```python
fo = open(fname, 'r')
txt = fo.read(2)
fo.seek(0)
for line in fo.readlines(): # 这个比较占用资源了，可能
    print(line)
fo.close()
```

```python
fo = open(fname, 'r')
txt = fo.read(2)
fo.seek(0) # 改变指针的位置
for line in fo: # 直接句柄，读大文件有效
    print(line)
fo.close()
```

## 写文件方法

- f.write(s)

- f.writelines(lines)

- f.seek(offset)

```python
fo = open("output.txt", "w+")
l = ["China", "American", "France"]
fo.writelines(l)
for line in fo:
    print(line)
fo.close()
```

```python
fo = open("output.txt", "r+")
for line in fo:
    print(line)
fo.close()
```
